Calcule usando Numpy:

1. **Análise Estatística:** Calcule a média, mediana e desvio padrão das temperaturas usando Numpy.
2. **Classificação de Temperaturas:** Classifique cada dia como "frio", "moderado" ou "quente". Você pode definir os limites para cada categoria. Por exemplo:
    - Frio: Temperatura abaixo de 5 graus.
    - Moderado: Temperatura entre 5 e 15 graus.
    - Quente: Temperatura acima de 15 graus.
3. **Dias Extremos:** Identifique os dias mais frios e mais quentes do ano.


```python
#Média, Mediana e Desvio Padrão

#Média
media = np.mean(temperaturas)
print("Média das temperaturas: ", media)

#Mediana
mediana = np.median(temperaturas)
print("Mediana das temperaturas: ", mediana)

#Desvio Padrão
desvio_padrao = np.std(temperaturas)
print("Desvio padrão das temperaturas: ", desvio_padrao)
```

```python
# Frio: Temperatura abaixo de 5 graus.
# Moderado: Temperatura entre 5 e 8 graus.
# Quente: Temperatura acima de 8 graus.

frio = temperaturas[(temperaturas <5)]
moderado = temperaturas[(temperaturas > 5) & (temperaturas < 8)]
quente = temperaturas[(temperaturas > 8)]

print("Frio:", frio)
print("Moderado:", moderado)
print("Quente:", quente)

#Classificação com a base completa:
classifica =  np.where(temperaturas < 5, "frio", np.where((temperaturas > 5)&(temperaturas<8) , "moderado", "quente"))
print(classifica)
```

```python
#Dias Extremos:
Dia_Max =  np.max(temperaturas)
Dia_Min = np.min(temperaturas)
print("Dia mais quente: ",Dia_Max)
print("Dia mais frio",Dia_Min)

Dias_frios = np.sort(temperaturas)[:3]
Dias_quentes = np.sort(temperaturas)[-3:]
print("Dias mais frios:" , Dias_frios)
print("Dias mais quentes:" , Dias_quentes)
```
