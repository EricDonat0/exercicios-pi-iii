```python
import matplotlib.pyplot as plt
import numpy as np

mes = np.arange(12) + 1 # meses de 1 a 12

# quantidades de produtos vendidos por mês. As entradas são de Janeiro a Dezembro
creme_facial = np.array([ 2500, 2630, 2140, 3400, 3600, 2760, 2980, 3700, 3540, 1990, 2340, 2900 ])
limpeza_facial = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])
pasta_dentaria = np.array([ 5200, 5100, 4550, 5870, 4560, 4890, 4780, 5860, 6100, 8300, 7300, 7400 ])
sabonete = np.array([ 9200, 6100, 9550, 8870, 7760, 7490, 8980, 9960, 8100, 10300, 13300, 14400 ])
shampoo = np.array([ 1200, 2100, 3550, 1870, 1560, 1890, 1780, 2860, 2100, 2300, 2400, 1800 ])
hidratante = np.array([ 1500, 1200, 1340, 1130, 1740, 1555, 1120, 1400, 1780, 1890, 2100, 1760 ])

# Gráfico 1 - Total de produtos Vendidos por mês - Linha
# Gráfico 2 - Gráfico com todos os produtos vendidos por mês - Linha
# Gráfico 3 - Comparativo de Creme Facial com Limpeza Facial por mês - Barras
# Gráfico 4 - Stack plot com o total de produtos vendidos a cada mês com todos os produtos
# Gráfico 5 - Pizza com % da quantidade produtos vendidos no ano em cada produto
```

# Gráfico 1 - Total de produtos Vendidos por mês - Linha

```python
num_mes = mes
produtos = creme_facial + limpeza_facial + pasta_dentaria + sabonete + shampoo + hidratante

fig, ax = plt.subplots()
ax.plot(num_mes, produtos)
```

# Gráfico 2 - Gráfico com todos os produtos vendidos por mês - Linha

```python
num_mes = mes

fig, ax = plt.subplots()

ax.plot(num_mes, creme_facial, color='green', label='Creme Facial')
ax.plot(num_mes, limpeza_facial, color='black', label='Limpeza Facial')
ax.plot(num_mes, pasta_dentaria, color='red', label='Pasta Dentária')
ax.plot(num_mes, sabonete, color='blue', label='Sabonete')
ax.plot(num_mes, shampoo, color='yellow', label='Shampoo')
ax.plot(num_mes, hidratante, color='pink', label='Hidratante')

ax.legend()
```

# Gráfico 3 - Comparativo de Creme Facial com Limpeza Facial por mês - Barras

```python
num_mes = mes

fig, ax = plt.subplots()

ax.bar(num_mes, creme_facial, width=0.7, edgecolor="black", linewidth=0.4, label='Creme Facial')
ax.bar(num_mes, limpeza_facial, width=0.7, edgecolor="blue", linewidth=0.4, label='Limpeza Facial')

ax.legend()
ax.set_title('Creme Facial X Limpeza Facial')
```

# Gráfico 4 - Stack plot com o total de produtos vendidos a cada mês com todos os produtos

```python
num_mes = mes

fig, ax = plt.subplots(figsize=(12,6))

ax.stackplot(num_mes, creme_facial, limpeza_facial, pasta_dentaria, sabonete, shampoo, hidratante,
             labels=["Creme Facial", "Limpeza Facial", "Pasta Dentária", "Sabonete", "Shampoo", "Hidratante"],
             colors=plt.get_cmap('tab10').colors)

ax.set_title('Vendas por mês')
ax.set_xlabel('Mês')
ax.set_ylabel('Quantidade Vendida')
ax.legend(loc='upper left')
ax.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

#  Gráfico 5 - Pizza com % da quantidade produtos vendidos no ano em cada produto

```python
produtos = [sum(creme_facial), sum(limpeza_facial), sum(pasta_dentaria),
            sum(sabonete), sum(shampoo), sum(hidratante)]
legenda = ["Creme Facial", "Limpeza Facial", "Pasta Dentária", "Sabonete", "Shampoo", "Hidratante"]

colors = plt.get_cmap('Blues')(np.linspace(0.3, 0.8, len(produtos)))  # tons de azul

fig, ax = plt.subplots(figsize=(8,6))

ax.pie(produtos, labels=None, colors=colors,
       wedgeprops={"linewidth": 0.5, "edgecolor": "white"}, autopct='%1.1f%%')

ax.legend(legenda, title="Produtos", loc="center left", bbox_to_anchor=(1, 0.5))
ax.set_title('Venda de cada produto no ano')

plt.show()
```
